# Visualization

Loading data first.

```python
model_dir = 'optlib4'
transfer = False
```

```python
import pandas as pd
optlib_raw = pd.read_csv('storage/{}_transfer/log.csv'.format(model_dir) if transfer else 'storage/{}/log.csv'.format(model_dir) )
optlib_mean = optlib_raw.rreturn_mean
optlib_lower = optlib_mean - optlib_raw.rreturn_std
optlib_upper = optlib_mean + optlib_raw.rreturn_std

vanilla_raw = pd.read_csv('storage/vanilla1_transfer/log.csv' if transfer else 'storage/vanilla1/log.csv')
vanilla_mean = vanilla_raw.rreturn_mean
vanilla_lower = vanilla_mean - vanilla_raw.rreturn_std
vanilla_upper = vanilla_mean + vanilla_raw.rreturn_std
```

plot using Plotly

```python
import plotly.graph_objects as go
import numpy as np
everyk = 200

cutoff = min(len(optlib_raw), len(vanilla_raw ))
x = list(optlib_raw['update'] if len(optlib_raw) >= cutoff else vanilla_raw['update'])
x = x[:cutoff][::everyk]

# Line 1
y1 = list(optlib_mean)[:cutoff][::everyk]
y1_upper = list(optlib_upper)[:cutoff][::everyk]
y1_lower = list(optlib_lower)[:cutoff][::everyk]

# Line 2
y2 = list(vanilla_mean)[:cutoff][::everyk]
y2_upper = list(vanilla_upper) [:cutoff][::everyk]
y2_lower = list(vanilla_lower) [:cutoff][::everyk]

x_rev = x[::-1]
y1_lower = y1_lower[::-1]
y2_lower = y2_lower[::-1]
```

```python

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x+x_rev,
    y=y1_upper+y1_lower,
    fill='toself',
    fillcolor='rgba(0,100,80,0.2)',
    line_color='rgba(255,255,255,0)',
    showlegend=False,
    name='Optlib (ours)',
))
fig.add_trace(go.Scatter(
    x=x+x_rev,
    y=y2_upper+y2_lower,
    fill='toself',
    fillcolor='rgba(0,176,246,0.2)',
    line_color='rgba(255,255,255,0)',
    name='Vanilla baseline',
    showlegend=False,
))
# fig.add_trace(go.Scatter(
#     x=x+x_rev,
#     y=y3_upper+y3_lower,
#     fill='toself',
#     fillcolor='rgba(231,107,243,0.2)',
#     line_color='rgba(255,255,255,0)',
#     showlegend=False,
#     name='Ideal',Rename
# ))

fig.add_trace(go.Scatter(
    x=x, y=y2,
    line_color='rgb(0,176,246)',
    name='Vanilla baseline',
))

fig.add_trace(go.Scatter(
    x=x, y=y1,
    line_color='rgb(0,100,80)',
    name='Optlib (ours)',
))

# fig.add_trace(go.Scatter(
#     x=x, y=y3,
#     line_color='rgb(231,107,243)',
#     name='Ideal',
# ))


fig.update_yaxes(
        title_text = "Mean returns",
        title_standoff = 10)

fig.update_xaxes(
        title_text = "Updates",
        title_standoff = 10)


fig.update_traces(mode='lines')
fig.show()
```
